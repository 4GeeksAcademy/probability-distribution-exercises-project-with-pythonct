## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Exercise 1

Write the above functions but for both normal and uniform distributions.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def dnorm(x, mean=0, sd=1):
    """
    Calculates the point estimate of the normal distribution
    """
    return norm.pdf(x, loc=mean, scale=sd)

def pnorm(q, mean=0, sd=1):
    """
    Calculates the cumulative probability of the normal distribution
    """
    return norm.cdf(q, loc=mean, scale=sd)

def qnorm(p, mean=0, sd=1):
    """
    Calculates the quantile function of the normal distribution
    """
    return norm.ppf(p, loc=mean, scale=sd)

def rnorm(n, mean=0, sd=1):
    """
    Generates random variables from the normal distribution
    """
    return norm.rvs(loc=mean, scale=sd, size=n)

# Example usage:
np.random.seed(42)

mean = 10
sd = 0.2

# Probability density at a point
print(f"Probability density at 2 for normal({mean}, {sd}): {dnorm(2, mean, sd)}")

# Cumulative probability up to a point
print(f"Cumulative probability less than 2 for normal({mean}, {sd}): {pnorm(2, mean, sd)}")

# Quantile function
print(f"Quantile at 0.9 for normal({mean}, {sd}): {qnorm(0.9, mean, sd)}")

# Generate random samples
random_samples = rnorm(2000, mean, sd)
print(f"Generated random samples from normal({mean}, {sd}): {random_samples[:10]}")

# Plotting
pmf = dnorm(random_samples, mean=mean, sd=sd)
cdf = pnorm(random_samples, mean=mean, sd=sd)
ppf = qnorm(np.linspace(0, 1, len(random_samples)), mean=mean, sd=sd)

fig, axis = plt.subplots(3, 1, figsize=(7, 10))

axis[0].plot(random_samples, pmf, 'r.')
axis[0].set_title('Probability Density Function')
axis[1].plot(random_samples, cdf, 'b.')
axis[1].set_title('Cumulative Distribution Function')
axis[2].plot(np.linspace(0, 1, len(random_samples)), ppf, 'g.')
axis[2].set_title('Quantile Function')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt

def duniform(x, a=0, b=1):
    """
    Calculates the probability density of the uniform distribution
    """
    return uniform.pdf(x, loc=a, scale=b-a)

def puniform(q, a=0, b=1):
    """
    Calculates the cumulative distribution function of the uniform distribution
    """
    return uniform.cdf(q, loc=a, scale=b-a)

def quniform(p, a=0, b=1):
    """
    Calculates the quantile function of the uniform distribution
    """
    return uniform.ppf(p, loc=a, scale=b-a)

def runiform(n, a=0, b=1):
    """
    Generates random variables from the uniform distribution
    """
    return uniform.rvs(loc=a, scale=b-a, size=n)

# Example usage:
np.random.seed(42)

a = 0  # Lower bound of the distribution
b = 10  # Upper bound of the distribution

# Probability density at a point
print(f"Probability density at 2 for uniform({a}, {b}): {duniform(2, a, b)}")

# Cumulative probability up to a point
print(f"Cumulative probability less than 2 for uniform({a}, {b}): {puniform(2, a, b)}")

# Quantile function
print(f"Quantile at 0.9 for uniform({a}, {b}): {quniform(0.9, a, b)}")

# Generate random samples
random_samples = runiform(2000, a, b)
print(f"Generated random samples from uniform({a}, {b}): {random_samples[:10]}")

# Plotting
pdf = duniform(random_samples, a=a, b=b)
cdf = puniform(random_samples, a=a, b=b)
ppf = quniform(np.linspace(0, 1, len(random_samples)), a=a, b=b)

fig, axis = plt.subplots(3, 1, figsize=(7, 10))

axis[0].plot(random_samples, pdf, 'r.')
axis[0].set_title('Probability Density Function')
axis[1].plot(random_samples, cdf, 'b.')
axis[1].set_title('Cumulative Distribution Function')
axis[2].plot(np.linspace(0, 1, len(random_samples)), ppf, 'g.')
axis[2].set_title('Quantile Function')

plt.tight_layout()
plt.show()
```
